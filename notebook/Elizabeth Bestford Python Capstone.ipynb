```python
plt.figure(figsize=(10, 10)) 
ax=sns.barplot(x='Total Amount of Fines in Dollars', y='State', data= barchart_top_ten, palette='colorblind')

plt.xlabel('Total Amount of Fines ($)')
plt.ylabel('State')
plt.title('Top 10 States with the Highest Fines for Nursing Homes')

for p in ax.patches:
    ax.text(p.get_width() + 500000, p.get_y() + p.get_height() / 2, f'${p.get_width():,.0f}', 
            horizontalalignment='left', verticalalignment='center', fontsize=10, color='black')
sns.despine()

plt.savefig('bar_chart_powerbi.png', bbox_inches='tight', dpi=300)

plt.show()


```

```python
us_population=pd.read_csv('../data/us_population_2024.csv',encoding='latin1')
us_population
```

```python
us_population.drop(index=0)
```

```python
us_population['table with row headers in column A and column headers in rows 3 through 4. (leading dots indicate sub-parts)']
```

```python
us_population.rename(columns={'table with row headers in column A and column headers in rows 3 through 4. (leading dots indicate sub-parts)':'US Population State', 'Unnamed: 6':'2024'}, inplace = True)
```

```python
population_columns=columns=['US Population State', '2024']
population_columns
```

```python
us_population_cleaned=us_population[population_columns]
us_population_cleaned
```

```python
us_population_cleaned=us_population_cleaned.dropna(subset='2024')
us_population_cleaned
```

```python
us_population_cleaned['US Population State'] = us_population_cleaned['US Population State'].str.lstrip('.')
us_population_cleaned
```

```python
us_population_cleaned.drop(index=range(2, 8), inplace=True)
```

```python
us_population_cleaned['2024'] = us_population_cleaned['2024'].replace({',': ''}, regex=True)
us_population_cleaned['2024'] = pd.to_numeric(us_population_cleaned['2024'], errors='coerce')
```

```python
population_sort=us_population_cleaned[['US Population State','2024']].nlargest(10, '2024')
population_sort
```

```python
state_abbr = {
    'California': 'CA', 'Texas': 'TX', 'Florida': 'FL', 'New York': 'NY', 
    'Pennsylvania': 'PA', 'Illinois': 'IL', 'Ohio': 'OH', 'Georgia': 'GA',
    'North Carolina': 'NC', 'Michigan': 'MI'
}
population_sort['US Population State'] = population_sort['US Population State'].replace(state_abbr)
```

```python
plt.figure(figsize=(10, 10))
ax = sns.barplot(x='2024', y='US Population State', data=population_sort, palette='colorblind')

plt.xlabel('US Population')
plt.ylabel('State')
plt.title('Top 10 States by Population in 2024')

for p in ax.patches:
    ax.text(p.get_width() + 50000, p.get_y() + p.get_height() / 2, f'{p.get_width():,.0f}', 
            horizontalalignment='left', verticalalignment='center', fontsize=10, color='black')

sns.despine()

plt.savefig('population_powerbi.png', bbox_inches='tight', dpi=300)

plt.show()
```
