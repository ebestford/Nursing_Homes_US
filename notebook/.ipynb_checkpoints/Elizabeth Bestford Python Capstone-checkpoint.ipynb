```python
import pandas as pd
```

```python
nursing_homes=pd.read_csv('../data/NH_capstone_data_CSV.csv')
nursing_homes
```

```python
nursing_homes=nursing_homes.rename(columns={'County/Parish': 'County','City/Town':'City', 'Qaulity Management (QM) Rating':'Quality Management (QM) Rating'})
#going to grab count of nursing homes per county per state???
```

```python
#nursing_homes.columns.tolist()
```

```python
nursing_homes[['Ownership Type']].value_counts()
```

```python
state_counts=nursing_homes[['State']].value_counts()
state_counts
```

```python
avg_nursing_homes=state_counts.mean()
avg_nursing_homes
#all of nursing home ownership types
```

```python
#wanting to make sure im pulling data with every possible FOR PROFIT combo
```

```python
state_for_profit=nursing_homes[nursing_homes['Ownership Type'].str.contains('For Profit', case=False, na=False)]
#state_ranking_for_profit
```

```python
state_count_for_profit=state_for_profit.groupby('State').size().sort_values(ascending=False)
state_count_for_profit
```

```python
avg_state_count_for_profit=state_count_for_profit.mean()
avg_state_count_for_profit
```

```python
#Same steps as above but not for NON PROFIT combos
```

```python
state_non_profit=nursing_homes[nursing_homes['Ownership Type'].str.contains('Non Profit', case=False, na=False)]
```

```python
state_count_non_profit=state_non_profit.groupby('State').size().sort_values(ascending=False)
state_count_non_profit
```

```python
avg_state_count_non_profit=state_count_non_profit.mean()
avg_state_count_non_profit
```

```python
county_state_count=nursing_homes.groupby(['State','County']).size().reset_index(name='Count')
county_state_count
#doesnt help too much, going to look at the top county for each state
```

```python
max_county=county_state_count.loc[county_state_count.groupby('State')['Count'].idxmax()]
max_county=max_county.sort_values(by='Count', ascending=False)
max_county
#this found us the top county in each state -- will pull in population data and circle back to this
```

```python
abuse_counts=nursing_homes.groupby(['State', 'Abuse Icon']).size().reset_index(name='Abuse Count')
```

```python
abuse_counts_pivot=abuse_counts_pivot.sort_values(by='Y', ascending=False)
```
